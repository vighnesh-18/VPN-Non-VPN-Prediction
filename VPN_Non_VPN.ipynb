```python
import pandas as pd
from scipy.io import arff
```

```python
arff_file_path = 'TimeBasedFeatures-Dataset-15s-VPN.arff'
```

```python
data, meta = arff.loadarff(arff_file_path)
```

```python
df=pd.DataFrame(data)
df.head()
```

```python
df.columns
```

```python
df['class1'].unique()
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df[df['class1']==b'Non-VPN'].shape, df[df['class1']==b'VPN'].shape
```

```python
df['class1'] = df['class1'].replace({b'Non-VPN': 0, b'VPN': 1})
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop('class1', axis=1), df['class1'], test_size=0.2, random_state=42)
```

```python
df['class1'].unique()
```

```python
import xgboost as xgb
model=xgb.XGBClassifier()
model.fit(x_train, y_train)
```

```python
import matplotlib.pyplot as plt
y_pred=model.predict(x_test)
y_pred
```

```python
import numpy as np

count_0 = np.count_nonzero(y_pred == 0)
print(f"Number of 0s in y_pred: {count_0}")
```

```python
count_1=y_pred.shape[0]-count_0
count_0,count_1
```

```python
from sklearn.metrics import classification_report,roc_auc_score
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Non-VPN', 'VPN'])
disp.plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)

# Plot
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve, average_precision_score

# Calculate precision-recall
precision, recall, _ = precision_recall_curve(y_test, y_pred)
ap = average_precision_score(y_test, y_pred)

# Plot
plt.plot(recall, precision, label=f'AP = {ap:.2f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.show()
```
